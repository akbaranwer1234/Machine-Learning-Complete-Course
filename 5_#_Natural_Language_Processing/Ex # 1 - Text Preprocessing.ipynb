### Text Cleaning

Text cleaning is a crucial step in Natural Language Processing (NLP) for several reasons:

- ***Data Preparation***: The goal of data preparation is to produce clean text that machines can analyze error-free. Cleaning is important for model building.

- ***Reduce Noise***: Text cleaning helps to get rid of unhelpful parts of the data, or noise. This includes removing typos, spelling mistakes, numbers, punctuations, emojis, emoticons, usernames, links, etc.

- ***Consistency***: Text cleaning ensures that the same word, whether it appears in different forms or in different parts of the text, is treated as the same word.

- ***Dimensionality Reduction***: It reduces the dimensionality of the data, which can improve the performance of NLP algorithms.

However, its important to note that text cleaning is task-specific and one needs to have a strong idea about what they want their end result to be. Therefore, its crucial to make this decision based on the context and requirements of your NLP task.

#### Removing Punctution Marks

```python
def remove_punctuation(text):
    # Define a string containing all the punctuation and special characters you want to remove
    punctuation_chars = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"

    # Initialize an empty string to store the cleaned text
    cleaned_text = ""

    # Iterate through each character in the input text
    for char in text:
        # Check if the character is not in the punctuation_chars
        if char not in punctuation_chars:
            cleaned_text += char

    return cleaned_text
```

```python
# Example Usage

input_text = "Hello, World! This is an example text without any punctuation."
cleaned_text = remove_punctuation(input_text)
```

```python
print(cleaned_text)
```

#### Removing Punctuation Marks Using Regular Expression (re) Library

The **`re`** library is a built-in module in Python that provides support for regular expressions. Regular expressions, often shortened as regex, are sequences of characters used to check whether a pattern exists in a given text (string) or not.
- Documentation: https://docs.python.org/3/library/re.html

The **`string`** module in Python is a built-in module that provides common string operations.
- Documentation: https://docs.python.org/3/library/string.html

```python
# Removing  Punctuation Marks/Special Characters
import re
import string

def remove_punctuation_and_special_chars(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove special characters
    text = re.sub('[^A-Za-z0-9 ]+', '', text)

    return text
```

```python
# Example Usage

text = "Hello, world! This is a test. @#"
clean_text = remove_punctuation_and_special_chars(text)
```

```python
print(clean_text)
```

#### Removing Urls

```python
# Removing URLs

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    text = url_pattern.sub(r'', text)
    return text
```

```python
# Example Usage

text = "This is a test. https://www.example.com is a URL."
clean_text = remove_urls(text)
```

```python
print(clean_text)
```

### Stopwords
Stopwords are words which, due to their high frequency in text, often dont offer significant insights on their own. They are usually filtered out before or after processing of natural language data (text). Examples of stopwords include the, and, is, in, etc

```python
def remove_stopwords(text):
    a= text.split()
    return " ".join([word for word in a if word not in stopwords.words('english')])
```

### Stemming
 Stemming is the process of reducing words to their base or root form, also known as the stem.

 For example, it can reduce words like chocolates, chocolatey, choco to the root word, **chocolate**, and retrieval, retrieved, retrieves reduce to the stem "**retrieve**". This process is used to normalize text and make it easier to process.

**PorterStemmer**  
The Porter Stemmer is based on a set of heuristics that are used to remove common suffixes from words. Its the most widely used stemming algorithm.

Some Alternatives to PorterStemmer in NLTK:

- *Snowball Stemmer*: Also known as the Porter2 Stemmer, it is an improvement over the original Porter Stemmer and supports stemming in several languages.

- *Lancaster Stemmer*: It is a more aggressive stemming algorithm than the Porter Stemmer. It follows a different approach and, in some cases, it might produce better results.


```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
ps = PorterStemmer()    ## Creating an object for porterstemmer
sno = SnowballStemmer('english')    ## Creating an object for snowballstemmer
ls = LancasterStemmer()   ## Creating an object for lancasterstemmer
example_words = ["earning","earned","earns", "transparent", "fairly", "bustling"]  ##Example words

for w in example_words:
    print(f"{w} - Porter: {ps.stem(w)}    Snowball: {sno.stem(w)}   Lancaster: {ls.stem(w)}")    ##Using ps object stemming the word
```

Its important to note that stemming can sometimes lead to over-stemming, where the stem produced is not a valid word or is not the correct root form of a word.

### Lemmatizer
Lemmatization usually refers to doing things properly with the use of vocabulary and morphological analysis of words, normally aiming to remove inflectional endings only and to return the base or dictionary form of a word, which is known as the lemma.

Lemmatization is different from Stemming. While both processes reduce a word to its base form, lemmatization takes into account the context of the word and produces a valid word, unlike stemming which may produce a non-word as the root form.

In simple words lemmatization does the same work as stemming, the difference is that lemmatization returns a meaningful word.

**Example:**  
*Stemming*  
history  histori  
*Lemmatizing*  
history  history

### Part of speech tagging
Part-of-Speech (POS) tagging, also known as grammatical tagging, is a process in corpus linguistics where each word in a text (or corpus) is marked as corresponding to a particular part of speech. This is based on both its definition and its context.

The main parts of speech are nouns, verbs, adjectives, adverbs, etc., but there are clearly many more categories and sub-categories. For example, different inflections of the same root word can get different parts of speech, resulting in a large number of tags.

POS tagging is an important step in natural language processing and understanding. It helps in understanding the context and the grammatical structure of sentences which can be useful in many applications like text-to-speech conversion, information retrieval, and machine translation.

In NLTK, the averaged_perceptron_tagger is a popular choice for part-of-speech tagging.

```python
nltk.download('averaged_perceptron_tagger')
```

The ***averaged_perceptron_tagger*** in NLTK is an implementation of the Averaged Perceptron algorithm, which is used for Part-of-Speech (POS) tagging.

The Averaged Perceptron is a type of perceptron algorithm. The main idea behind it is to take the average of all the weights of the perceptron throughout the learning process. This helps to stabilize the final weights and makes them less sensitive to the last few training examples.

The averaged_perceptron_tagger takes a sentence as input and returns a list of tuples, where each word in the sentence is paired with a part of speech tag. The tagger is trained on a labeled dataset, where each word in a sentence is already tagged with a part of speech.

```python
sample_text = '''
An sincerity so extremity he additions. Her yet there truth merit.
Mrs all projecting favourable now unpleasing. Son law garden chatty temper.
Oh children provided to mr elegance marriage strongly. Off can admiration
prosperous now devonshire diminution law.
'''
from nltk.tokenize import word_tokenize
words = word_tokenize(sample_text)
print(nltk.pos_tag(words))
```

###### Some Alternatives to the Average Perceptron Tagger  
- *UnigramTagger*: The UnigramTagger is a simple statistical tagging algorithm that works by assigning the most likely tag to each token.

- *BigramTagger*: The BigramTagger is a slightly more advanced statistical tagging algorithm that works by assigning tags based on the context of the word in the sentence.

- *TrigramTagger*: The TrigramTagger is similar to the BigramTagger, but it considers the two words before it in the sentence, rather than just one.

- *HiddenMarkovModelTagger*: The HiddenMarkovModelTagger is a probabilistic sequence model that assigns tags based on the sequence of words in the sentence.


```python
nltk.download('treebank')
```

The treebank in NLTK refers to the Penn Treebank. The Penn Treebank is one of the most widely used corpora for the evaluation of models for sequence labelling. Its a large annotated corpus of English, produced by the University of Pennsylvania.


the-penn-treebank-a-syntactially-annotated-corpus-l.jpg

```python
from nltk.tag import UnigramTagger, BigramTagger, TrigramTagger, HiddenMarkovModelTagger
from nltk.corpus import treebank

train_sents = treebank.tagged_sents()[:3000]
tagger = UnigramTagger(train_sents)

tokens = sample_text.split()
print(tagger.tag(tokens))

tagger = BigramTagger(train_sents)
print(tagger.tag(tokens))

tagger = TrigramTagger(train_sents)
print(tagger.tag(tokens))

tagger = HiddenMarkovModelTagger.train(train_sents)
print(tagger.tag(tokens))
```

Remember, these alternatives might not be as comprehensive as averaged_perceptron_tagger and their effectiveness can vary based on your specific use case. Its always a good idea to experiment with different tools and see which one works best for your needs.

### Word Clouds

A word cloud, also known as a text cloud or tag cloud, is a form of visual representation. It is a grouping of words that are displayed in various sizes: the larger and bolder the term, the more frequently it appears in a document and the more important it is. Word clouds include data visualization, text data, font colors, word frequency analysis, and specific word graphics. They are ideal techniques to extract the most relevant sections of textual material, from blog posts to databases.

In essence, word clouds are a powerful way to visualize what your audience really thinks about a topic. They are easy to read, quick to produce, and simple to understand.

**Dataset:** LinkedIn Job Postings - 2023

**Kaggle Link:** https://www.kaggle.com/datasets/arshkon/linkedin-job-postings

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

titles = ' '.join(df['title']).lower()

# Load a custom shape for the word cloud (e.g., a heart shape image)
custom_shape = np.array(Image.open("cloud.png"))

# Create word clouds for spam and ham messages in the custom shape
def generate_word_cloud(text, custom_shape, title):
    wordcloud = WordCloud(width=800, height=800, background_color='white', mask=custom_shape).generate(text)

    plt.figure(figsize=(8, 8))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(title)
    plt.show()

# Generate word clouds for ham
generate_word_cloud(titles, custom_shape, "Word Cloud for Job Titles")
```
