Simulating Coin Flips:

Create a function to simulate a coin flip. You can represent "heads" as 1 and "tails" as 0. Use a loop
to simulate a series of coin flips (e.g., 100 flips). Calculate the probability of getting heads or tails in
the simulation. Analyze the effect of increasing the number of flips on probability. Visualize the
results with a bar chart using Matplotlib

```python
def coinFlipSimulation (numberOfFlips):

    totalHeads = 0
    totalTails = 0

    flipHistory = []
    # flipHistoryEventWise = []

    for i in range(numberOfFlips):

        flipResult = random.randint(0, 1) # 1 for heads, 0 for tails
        flipHistory.append(flipResult)

        if flipResult == 0:
            totalTails += 1
        else:
            totalHeads += 1 

        # allHeads = len(flipHistory[flipHistory == 1])
        # allTails = len(flipHistory[flipHistory == 0])

        # flipHistoryEventWise.append([allHeads/i, allTails/i])
    
    headsProbability =  totalHeads/numberOfFlips
    tailsProbability = totalTails/numberOfFlips

    return flipHistory, headsProbability, tailsProbability #, flipHistoryEventWise
```

Simulating Dice Rolls

Create a function to simulate a dice roll. You can use the random module to generate random
numbers between 1 and 6. Use a loop to simulate multiple dice rolls (e.g., 100 rolls of a fair 6-sided
die). Calculate the probability of each outcome (1 through 6) in the simulation. Analyze the effect of
increasing the number of rolls on probability. Visualize the results with a bar chart using Matplotlib.

```python
def diceRollSimulation (numberOfRolls):

    one = 0
    two = 0
    three = 0
    four = 0
    five = 0
    six = 0

    diceHistory = []

    for i in range(numberOfRolls):

        diceResult = random.randint(1, 6)
        diceHistory.append(diceResult)

        if diceResult == 1:
            one += 1
        elif diceResult == 2:
            two += 1
        elif diceResult == 3:
            three += 1
        elif diceResult == 4:
            four += 1
        elif diceResult == 5:
            five += 1
        else:
            six += 1 
    
    onesProbability =  one/numberOfRolls
    towsProbability = two/numberOfRolls
    threesProbability = three/numberOfRolls
    foursProbability = four/numberOfRolls
    fivesProbability = five/numberOfRolls
    sixProbability = six/numberOfRolls

    return diceHistory, onesProbability, towsProbability, threesProbability, foursProbability, fivesProbability, sixProbability
```

```python
numberOfRolls = 100

diceHistory, onesProbability, towsProbability, threesProbability, foursProbability, fivesProbability, sixProbability = diceRollSimulation(numberOfRolls)

label = ["One", "Two", "Three", "Four", "Five", "Six"]
probabilities = [onesProbability, towsProbability, threesProbability, foursProbability, fivesProbability, sixProbability]
plt.bar(label, probabilities)
plt.ylabel("Probability")
plt.title(f"Dice Rolled simulation on {numberOfRolls} flips")
```

```python
def calculateFactorial (number):
    
    if number == 1 or number == 2:
        return number

    return number*calculateFactorial(number-1)
```

```python
def calculateNCR (n, r):

    if r == 0:
        return 1
    
    return calculateFactorial(n)/(calculateFactorial(r)*calculateFactorial(n-r))
    
```

```python
def calculateMVSD (n, p):

    q = 1-p

    mean = n*p
    variance = n*p*q
    standardDeviation = sqrt(variance)

    return mean, variance, standardDeviation
```

```python
import random

def generate_binomial_samples(n, p, size):
    """
    Generate random samples from a binomial distribution.

    Args:
        n (int): Number of trials.
        p (float): Probability of success in each trial.
        size (int): Number of samples to generate.

    Returns:
        list: A list of generated samples.
    """
    samples = []
    for _ in range(size):
        # Simulate one binomial trial
        trial_result = sum(1 for _ in range(n) if random.random() < p)
        # print(trial_result)
        samples.append(trial_result)
    return samples

# Example usage:
n = 10  # Number of trials
p = 0.3  # Probability of success in each trial
size = 10000  # Number of samples to generate

binomial_samples = generate_binomial_samples(n, p, size)

# Print the first 10 samples
print(binomial_samples[:10])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# 1. Binomial Distribution
n = 10  # Number of trials
p = 0.3  # Probability of success
binomial_samples = np.random.binomial(n, p, 10000)  # Generate 10,000 samples

# Plot histogram
plt.hist(binomial_samples, bins=range(0, n + 2), density=True, alpha=0.6, color='b', rwidth=0.85)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()

# Calculate mean and variance
binomial_mean = np.mean(binomial_samples)
binomial_var = np.var(binomial_samples)

# 2. Poisson Distribution
lambda_ = 3  # Average rate
poisson_samples = np.random.poisson(lambda_, 10000)  # Generate 10,000 samples

# Plot histogram
plt.hist(poisson_samples, bins=range(0, 15), density=True, alpha=0.6, color='g', rwidth=0.85)
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()

# Calculate mean and variance
poisson_mean = np.mean(poisson_samples)
poisson_var = np.var(poisson_samples)

# 3. Normal Distribution
mu = 2  # Mean
sigma = 1.5  # Standard deviation
normal_samples = np.random.normal(mu, sigma, 10000)  # Generate 10,000 samples

# Plot histogram
plt.hist(normal_samples, bins=30, density=True, alpha=0.6, color='r', rwidth=0.85)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()

# Calculate mean and variance
normal_mean = np.mean(normal_samples)
normal_var = np.var(normal_samples)

# 4. Exponential Distribution
lambda_ = 0.5  # Rate parameter (inverse of mean)
exponential_samples = np.random.exponential(1 / lambda_, 10000)  # Generate 10,000 samples

# Plot histogram
plt.hist(exponential_samples, bins=30, density=True, alpha=0.6, color='y', rwidth=0.85)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()

# Calculate mean and variance
exponential_mean = np.mean(exponential_samples)
exponential_var = np.var(exponential_samples)

# Compare means and variances to theoretical values
print(f'Binomial Mean (Theoretical vs. Sample): {n * p} vs. {binomial_mean}')
print(f'Binomial Variance (Theoretical vs. Sample): {n * p * (1 - p)} vs. {binomial_var}')
print(f'Poisson Mean (Theoretical vs. Sample): {lambda_} vs. {poisson_mean}')
print(f'Poisson Variance (Theoretical vs. Sample): {lambda_} vs. {poisson_var}')
print(f'Normal Mean (Theoretical vs. Sample): {mu} vs. {normal_mean}')
print(f'Normal Variance (Theoretical vs. Sample): {sigma**2} vs. {normal_var}')
print(f'Exponential Mean (Theoretical vs. Sample): {1 / lambda_} vs. {exponential_mean}')
print(f'Exponential Variance (Theoretical vs. Sample): {1 / lambda_**2} vs. {exponential_var}')
```
