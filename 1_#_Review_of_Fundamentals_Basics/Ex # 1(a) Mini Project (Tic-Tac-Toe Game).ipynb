 Mini Project: Tic-Tac-Toe Game with AI Player:

```python
print("Tic-Tac-Toe game Mini project with AI character by")
print("Muhammad Akbar")

# Initialize the Tic-Tac-Toe board
board = [' ' for _ in range(9)]

# Function to print the Tic-Tac-Toe board
def print_board():
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--|---|--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--|---|--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
    print('')

# Function to check if the board is full (a tie)
def is_board_full(board):
    return ' ' not in board

# Function to check if a player has won
def check_win(board, player):
    win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6)]
    for condition in win_conditions:
        if board[condition[0]] == board[condition[1]] == board[condition[2]] == player:
            return True
    return False

# Function to make a player's move
def make_move(board, position, player):
    if board[position] == ' ':
        board[position] = player

# Minimax Algorithm for the computer player (I just copy and paste this because i am not good in this algorithm but trying to learn it)
def minimax(board, depth, maximizing_player):
    if check_win(board, 'O'):
        return 1
    if check_win(board, 'X'):
        return -1
    if is_board_full(board):
        return 0

    if maximizing_player:
        max_eval = -float('inf')
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'
                eval = minimax(board, depth + 1, False)
                board[i] = ' '
                max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'
                eval = minimax(board, depth + 1, True)
                board[i] = ' '
                min_eval = min(min_eval, eval)
        return min_eval

# Function to make the computer player's move using minimax (I just copy and paste this because i am not good in this algorithm but trying to learn it)
def computer_move(): 
    best_move = -1
    best_eval = -float('inf')
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            eval = minimax(board, 0, False)
            board[i] = ' '
            if eval > best_eval:
                best_eval = eval
                best_move = i
    return best_move

# Main game loop
while True:
    print_board()
    
    # Human player's turn
    while True:
        try:
            user_move = int(input("Enter your move (1-9): ")) - 1
            if user_move >= 0 and user_move < 9 and board[user_move] == ' ':
                make_move(board, user_move, 'X')
                print_board()
                break
            else:
                print("Invalid move. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    # Check if the human player wins
    if check_win(board, 'X'):
        print_board()
        print("You win!")
        break

    # Check for a tie
    if is_board_full(board):
        print_board()
        print("It's a tie!")
        break

    # Computer player's turn
    computer_choice = computer_move()
    make_move(board, computer_choice, 'O')
    # print_board()
    

    # Check if the computer player win
    if check_win(board, 'O'):
        print_board()
        print("Computer wins!")
        break
```
