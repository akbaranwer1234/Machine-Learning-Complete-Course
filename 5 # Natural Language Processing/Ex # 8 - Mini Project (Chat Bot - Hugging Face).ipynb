```python
repo_id = 'mistralai/Mistral-7B-Instruct-v0.1'
llm = HuggingFaceHub(
    repo_id=repo_id, model_kwargs={"max_new_toke": 1024, "temperature": 0.5}
)
```

```python
system_propmpt = """You are a chatbot which helps the founders and enterpreneurs,
                to answer business related queries.
                If a user asks question other than business,
                ask them to only ask question and enterpreneurship.
                Given the Context and chat history answer the user question."""

B_INST, E_INST = "<s>[INST] ", " [/INST]"
template = (

            B_INST
            + system_propmpt
            + """

      Context: {chat_history} \n{context}
      User: {question}"""
            + E_INST
    )

prompt = PromptTemplate(input_variables=['context', 'chat_history', 'question'], template=template)
```

```python
from langchain.memory import ConversationBufferMemory
```

```python
memory = ConversationBufferMemory()
# memory.save_context({"input": "hi"}, {"output": "whats up"})
```

```python
memory = ConversationBufferMemory(memory_key='chat_history', return_messages=True, output_key='answer')
```

```python
retriever = db.as_retriever()
```

```python
chain = ConversationalRetrievalChain.from_llm(
    llm = llm,
    retriever = db.as_retriever(),
    memory = memory,
    # return_generated_question = False,
    # rephrase_question = True,
    # return_source_documents= False
)
```

```python
result = chain('What are the key elements of business growth plan?')
result['answer']
```

```python
result = chain('What about the fourth element?')
result['answer']
```

```python
result = chain('What are the targeted element?')
result['answer']
```

```python
result = chain('What are the targeted element? does it also belong to key elements of business growth plan?')
result['answer']
```

```python
import gradio as gr
```

```python
def chatllm(user_msg, history):
  result = chain(user_msg)
  return result['answer']

demo = gr.ChatInterface(fn=chatllm,)
demo.launch()
```
