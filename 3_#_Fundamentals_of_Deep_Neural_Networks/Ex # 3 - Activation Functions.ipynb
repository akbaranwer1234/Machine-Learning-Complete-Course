```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid (x):
    return 1/(1 + np.exp(-x))

def relu (x):
    return np.maximum(0, x)

def tanh (x):
    return np.tanh(x)

def softmax (x):
    return np.exp(x) / np.sum(np.exp(x))

def softplus (x):
    return np.log(1 + np.exp(x))

def leaky_relu (x):
    return np.maximum(0.1 * x, x)

def parametric_relu (x, alpha):
    return np.maximum(alpha * x, x)

def selu (x, alpha, lambd):
    return lambd * np.where(x < 0, alpha * (np.exp(x) - 1), x)

def gaussian (x, mu, sigma):
    return np.exp(-np.power(x - mu, 2) / (2 * np.power(sigma, 2)))

def elu (x, alpha):
    return np.where(x < 0, alpha * (np.exp(x) - 1), x)

def swish (x, beta):
    return x * sigmoid(beta * x)

def signc (x):
    return np.sinc(x)
```

```python
x = np.linspace(-5, 5, 1000)
```

```python
sigmoid_output = sigmoid(x)
relu_output = relu(x)
tanh_output = tanh(x)
softmax_output = softmax(x)
softplus_output = softplus(x)
leaky_relu_output = leaky_relu(x)
parametric_relu_output = parametric_relu(x, 0.01)
selu_output = selu(x, 0.01, 0.5)
gaussian_output = gaussian(x, 0.01, 0.51)
elu_output = elu(x, 0.01)
swish_output = swish(x, 0.51)
signc_output = signc(x)
```

```python
plt.figure(figsize=(30, 45))
plt.subplot(5, 3, 1)
plt.plot(x, sigmoid_output, label='Sigmoid')
plt.title('Sigmoid Activation')

plt.subplot(5, 3, 2)
plt.plot(x, relu_output, label='Relu')
plt.title('Relu Activation')

plt.subplot(5, 3, 3)
plt.plot(x, tanh_output, label='Tanh')
plt.title('Tanh Activation')

plt.subplot(5, 3, 4)
plt.plot(x, softmax_output, label='Softmax')
plt.title('Softmax Activation')

plt.subplot(5, 3, 5)
plt.plot(x, softplus_output, label='Softplus')
plt.title('Softplus Activation')

plt.subplot(5, 3, 6)
plt.plot(x, leaky_relu_output, label='Leaky Relu')
plt.title('Leaky Relu Activation')

plt.subplot(5, 3, 7)
plt.plot(x, parametric_relu_output, label='Parametric Relu')
plt.title('Parametric Relu Activation')

plt.subplot(5, 3, 8)
plt.plot(x, selu_output, label='Selu')
plt.title('Selu Activation')

plt.subplot(5, 3, 9)
plt.plot(x, gaussian_output, label='Gaussian')
plt.title('Gaussian Activation')

plt.subplot(5, 3, 10)
plt.plot(x, elu_output, label='Elu')
plt.title('Elu Activation')

plt.subplot(5, 3, 11)
plt.plot(x, swish_output, label='Swish')
plt.title('Swish Activation')

plt.subplot(5, 3, 12)
plt.plot(x, signc_output, label='Signc')
plt.title('Signc Activation')
```
