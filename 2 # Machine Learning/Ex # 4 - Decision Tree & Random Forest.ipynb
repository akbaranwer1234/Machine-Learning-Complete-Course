```python
import pandas as pd
import numpy as np
```

```python
dataPath = '/titanic.csv'
df = pd.read_csv(dataPath)
df.head(5)
```

```python
df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1, inplace = True)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'], test_size = 0.2, random_state = 42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
transformer1 = ColumnTransformer([
    ('impute_age', SimpleImputer(), [2]),
    ('impute_embarked', SimpleImputer(strategy = 'most_frequent'), [6]),
], remainder='passthrough')

transformer2 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [1, 6])
])

# transformer3 = ColumnTransformer([
#     ('scale', MinMaxScaler(), slice(0, 10))
# ])

transformer4 = ColumnTransformer([
    ('standard_scale', StandardScaler(), slice(0, 10))
])
```

```python
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto', 0.1, 1],
}

# Create a grid search object
svm_classifier = SVC()
transformer5 = GridSearchCV(svm_classifier, param_grid, cv=5)
```

```python
transformer6 = LogisticRegression()
```

```python
transformer7 = DecisionTreeClassifier()
```

```python
transformer8 = CountVectorizer()
```

```python
transformer9 = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipeSVC = Pipeline([
    ('Transformer1', transformer1),
    ('Transformer2', transformer2),
    # ('Transformer3', transformer3),
    ('Transformer4', transformer4),
    ('Transformer5', transformer5),
])
```

```python
pipeLogistic = Pipeline([
    ('Transformer1', transformer1),
    ('Transformer2', transformer2),
    # ('Transformer3', transformer3),
    ('Transformer4', transformer4),
    ('Transformer6', transformer6),
])
```

```python
pipeDecision = Pipeline([
    ('Transformer1', transformer1),
    ('Transformer2', transformer2),
    # ('Transformer3', transformer3),
    ('Transformer4', transformer4),
    ('Transformer7', transformer7),
])
```

```python
pipeNaive = Pipeline([
    ('Transformer1', transformer1),
    ('Transformer2', transformer2),
    # ('Transformer3', transformer3),
    ('Transformer4', transformer4),
    ('Transformer8', transformer8),
])
```

```python
pipeRandom = Pipeline([
    ('Transformer1', transformer1),
    ('Transformer2', transformer2),
    # ('Transformer3', transformer3),
    ('Transformer4', transformer4),
    ('Transformer9', transformer9),
])
```

## SVC model

```python
pipeSVC.fit(X_train, y_train)
```

```python
y_predSVC = pipeSVC.predict(X_test)
y_predSVC
```

```python
# Evaluate the classifier
accuracy = accuracy_score(y_test, y_predSVC)
confusion = confusion_matrix(y_test, y_predSVC)
classification_rep = classification_report(y_test, y_predSVC)

# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", classification_rep)
```

## Logistic regression

```python
pipeLogistic.fit(X_train, y_train)
```

```python
y_predLogistic = pipeLogistic.predict(X_test)
y_predLogistic
```

```python
accuracy = accuracy_score(y_test, y_predLogistic)
confusion = confusion_matrix(y_test, y_predLogistic)
classification_rep = classification_report(y_test, y_predLogistic)

# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", classification_rep)
```
